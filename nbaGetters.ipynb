```python
#Search for a player id by name
def getPlayerId(name):
    try:
        nba_players = players.get_players()
        person = [player for player in nba_players
                           if player['full_name'].lower() == name.lower()][0]
        return person['id']
    except:
        print("Player Not Found")
        return None
```

```python
#Get Career stats by player name

def getCareerStats(name):
    playerId = getPlayerId(name)
    
    if playerId == None:
        return None
    else:
        career_info = playercareerstats.PlayerCareerStats(player_id=playerId)
        career_info = career_info.get_normalized_dict()
        career_info = career_info.get('CareerTotalsRegularSeason')[0]

        ppg = (career_info.get('PTS') / career_info.get('GP'))
        reb = (career_info.get('REB') / career_info.get('GP'))
        ast = (career_info.get('AST') / career_info.get('GP'))

        stats = [round(ppg,1),round(reb,1), round(ast,1)]

        return(stats)
```

```python
#search for player in file
def playerFinder(name,filename):
    f = open(filename)
    for l in f:
        line = l.split(",")
        if line[0].lower() == name.lower():
            return [name.lower(),line[1].rstrip('\n')]
    return 0
```

```python
#Get number of rings for entered player
def getRings(name):
    return playerFinder(name,'rings.txt')
```

```python
#Get number of all defensive teams for entered player
def getAllDefense(name):
    return playerFinder(name,'all-Defense.txt')
```

```python
#Get number of all-nba selections for entered player
def getAllNBA(name):
    return playerFinder(name,'all-NBA.txt')
```

```python
#Get number of times entered player has won an award
def getAwardNumber(name,filename):
    number = 0
    names = []
    f = open(filename,'r')
    for line in f:
        if line.lower() == name.lower()+'\n':
            number+=1
    return number
```

```python
#Get number of DPOY for entered player
def getDPOY(name):
    return [name, getAwardNumber(name,'dpoy.txt')]
```

```python
#Get Number of all-star appearences for entered player
def getAllStars(name):
    return playerFinder(name, 'allStars.txt')
```

```python
def getScoringTitles(name):
    return [name, getAwardNumber(name,'scoringTitle.txt')]
```

```python
#Get number of finals mvps for entered player
def getFinalsMVPs(name):
    return [name, getAwardNumber(name,'finalsMVP.txt')]
```

```python
from nba_api.stats.endpoints import commonplayerinfo
from nba_api.stats.static import teams
from nba_api.stats.static import players
#from nba_api.stats.endpoints import playerheadlinestats
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.endpoints import playerdashboardbyyearoveryear
from nba_api.stats.endpoints import playerdashboardbygeneralsplits

class Player():
    def __init__(self):
        pass
    
    def getPlayerId(self, name):
        try:
            nba_players = players.get_players()
            person = [player for player in nba_players
                               if player['full_name'].lower() == name.lower()][0]
            return person['id']
        except:
            print("Player Not Found")
            return None
    
    def getName(self, name):
        nba_players = players.get_players()
        player = [player for player in nba_players
                       if player['full_name'].lower() == name][0]
        playerId = player['id']
        player_info = commonplayerinfo.CommonPlayerInfo(player_id=playerId)
        player_info = player_info.get_normalized_dict()
        player_info = player_info.get('PlayerHeadlineStats')[0] 
        player_info = player_info.get('PLAYER_NAME')
        return player_info

    def getCareerStats(self, name):
        playerId = self.getPlayerId(name)
    
        if playerId == None:
            return None
        else:
            career_info = playercareerstats.PlayerCareerStats(player_id=playerId)
            career_info = career_info.get_normalized_dict()
            career_info = career_info.get('CareerTotalsRegularSeason')[0]

            ppg = (career_info.get('PTS') / career_info.get('GP'))
            reb = (career_info.get('REB') / career_info.get('GP'))
            ast = (career_info.get('AST') / career_info.get('GP'))

            stats = [str(round(ppg,1))+'ppg ',str(round(reb,1))+'rpg ', str(round(ast,1))+'apg']

            return(stats)

    
    def playerFinder(self, name, filename):
        f = open(filename)
        for l in f:
            line = l.split(",")
            if line[0].lower() == name.lower():
                return [name.lower(),line[1].rstrip('\n')]
        return [name.lower(), '0']
    
    def getAllStars(self, name):
        return self.playerFinder(name, 'allStars.txt')
    
    def getAwardNumber(self, name,filename):
        number = 0
        names = []
        f = open(filename,'r')
        for line in f:
            if line.lower() == name.lower()+'\n':
                number+=1
        return number
    
    def getAllNBA(self, name):
        return self.playerFinder(name,'all-NBA.txt')
    
    
    def getAllDefense(self, name):
        return self.playerFinder(name,'all-Defense.txt')
    
    
    def getScoringTitles(self,name):
        return [name, self.getAwardNumber(name,'scoringTitle.txt')]
    
    
    def getRings(self,name):
        return self.playerFinder(name,'rings.txt')
    
    
    def getMVPs(self,name):
        return self.getAwardNumber(name, 'mvp.txt')
    
    def getFinalsMVPs(self, name):
        return [name, str(self.getAwardNumber(name,'finalsMVP.txt'))]
    
    def getDPOY(self, name):
        return [name, self.getAwardNumber(name,'dpoy.txt')]
    
    def getAll(self, name):
        allStars = self.getAllStars(name)
        championships = self.getRings(name)
        fMvps = self.getFinalsMVPs(name)
        alldefense = self.getAllDefense(name)
        allnba = self.getAllNBA(name)
        scoringT = self.getScoringTitles(name)
        dpoy = self.getDPOY(name)
        mvp = self.getMVPs(name)
        return [int(allStars[1]), int(championships[1]), int(fMvps[1]), int(alldefense[1]), int(allnba[1]), int(scoringT[1]), int(dpoy[1])]

    def goatScore(self, name):
        favStats = self.getAll(name)
        score = (favStats[0]/20 + favStats[1]/8 + favStats[2]/6 + favStats[3]/18 +favStats[4]/16 + favStats[5]/10 + favStats[6]/3)*(100/7)
        return round(score,2)
    
```

```python
import customtkinter
class App(customtkinter.CTk):
    lines = []
    lines2 = []
    created = 0
    

    def __init__(self):
        super().__init__()

        self.title("NBA Accolades Comparer")
        self.geometry("750x500")

        self.openingFrame = customtkinter.CTkFrame(self)
        self.openingFrame.pack(pady =12, padx =10, fill="both", expand = True)
        
        self.label = customtkinter.CTkLabel(master=self.openingFrame, text= "NBA Accolades Comparer", font=("Roboto", 24))
        self.label.pack(side = 'top', pady=12, padx = 10)
        
        self.entry1 = customtkinter.CTkEntry(master= self.openingFrame, placeholder_text="Enter Player 1 Name")
        self.entry1.place( x=225,y=60)

        self.entry2 = customtkinter.CTkEntry(master= self.openingFrame, placeholder_text="Enter Player 2 Name")
        self.entry2.place(x=375,y=60)
        
        self.search = customtkinter.CTkButton(master=self.openingFrame, text= "Search", command =self.search)
        self.search.pack(side = 'bottom', pady=12, padx =10)
        
        #Create text lines for each stat
        self.line1 = customtkinter.CTkLabel(master=self.openingFrame, text=' ')
        self.line2 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line3 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line4 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line5 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line6 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line7 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line8 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line9 = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.goatScore = customtkinter.CTkLabel(master=self.openingFrame, text='') 
        
        
        self.line1b = customtkinter.CTkLabel(master=self.openingFrame, text='')
        self.line2b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line3b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line4b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line5b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line6b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line7b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line8b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.line9b = customtkinter.CTkLabel(master=self.openingFrame, text= '')
        self.goatScore2 = customtkinter.CTkLabel(master=self.openingFrame, text='')
        
    
    def createIcons(self):
        from PIL import Image
        self.my_image = customtkinter.CTkImage(dark_image=Image.open("greenSqr.png"), size=(5, 5))
        
        counter = 0
        while counter < 8:
            App.lines.append(customtkinter.CTkLabel(master=self.openingFrame, image=self.my_image, text=""))
            counter+=1

        counter2 = 0
        while counter2 < 8:
            App.lines2.append(customtkinter.CTkLabel(master=self.openingFrame, image=self.my_image, text=""))
            counter2+=1
        App.created = 1
        return [App.lines, App.lines2]
      
    def deleteIcons(self):
        i = 0
        while i < 8:
            App.lines[i].destroy()
            App.lines2[i].destroy()
            i+=1
        
        i = 0
        while i < 8:
            App.lines.pop()
            App.lines2.pop()
            i+=1
        
    def search(self):
            
        if App.created == 1:
            self.deleteIcons()
            App.created = 0
        
            
        
        player = Player()
        axis = 120
        
        #Player 1 full stats
        playerName = self.entry1.get().lower()
        
        stats = player.getCareerStats(playerName)
        
        self.line1.place( x = axis, y= 120)
        self.line1.configure(text='Name: '+player.getName(playerName))
        
        self.line2.place( x = axis, y= 150)
        self.line2.configure(text= 'Career stats: '+stats[0]+stats[1]+stats[2])
        
        self.line3.place( x = axis, y= 180)
        self.line3.configure(text='All-Star Appearances: '+player.getAllStars(playerName)[1])
        
        player1AllStars = int(player.getAllStars(playerName)[1])
        
        self.line4.place( x = axis, y= 210)
        self.line4.configure(text='All-NBA Selections: '+player.getAllNBA(playerName)[1])
        
        player1AllNBA = int(player.getAllNBA(playerName)[1])
        
        self.line5.place( x = axis, y= 240)
        self.line5.configure(text='All-Defensive Selections: '+player.getAllDefense(playerName)[1])
        
        player1AllDefense = int(player.getAllDefense(playerName)[1])
        
        self.line6.place( x = axis, y= 270)
        self.line6.configure(text= 'Championships: '+player.getRings(playerName)[1])
        
        player1GetRings = int(player.getRings(playerName)[1])
        
        self.line7.place( x = axis, y= 300)
        self.line7.configure(text= 'Finals MVPs: '+player.getFinalsMVPs(playerName)[1])
        
        player1FinalsMvps = int(player.getFinalsMVPs(playerName)[1])
        
        self.line8.place( x = axis, y= 330)
        self.line8.configure(text= 'Regular Season MVPs: '+str(player.getMVPs(playerName)))
        
        player1Mvp = int(player.getMVPs(playerName))
        
        self.line9.place( x = axis, y= 360)
        self.line9.configure(text= 'Scoring Titles: '+str(player.getScoringTitles(playerName)[1]))
        
        player1ScoringTitles = int(player.getScoringTitles(playerName)[1])
        
        self.goatScore.place( x = axis, y= 385)
        self.goatScore.configure(text= 'GOAT score: '+str(player.goatScore(playerName)))
        self.goatScore.configure(text_color = 'light green')
        self.goatScore.configure(font =('Roboto',15))
        
        player1GoatScore = int(player.goatScore(playerName))
        
        
        #Player 2 full stats
        playerName2 = self.entry2.get().lower()
        axis2 = 400
        stats2 = player.getCareerStats(playerName2)
        
        self.line1b.place( x = axis2, y= 120)
        self.line1b.configure(text = 'Name: '+player.getName(playerName2))
        
        self.line2b.place( x = axis2, y= 150)
        self.line2b.configure(text= 'Career stats: '+stats2[0]+stats2[1]+stats2[2])
        
        self.line3b.place( x = axis2, y= 180)
        self.line3b.configure(text= 'All-Star Appearances: '+player.getAllStars(playerName2)[1])
        
        player2AllStars = int(player.getAllStars(playerName2)[1])
        
        self.line4b.place( x = axis2, y= 210)
        self.line4b.configure(text= 'All-NBA Selections: '+player.getAllNBA(playerName2)[1])
        
        player2AllNBA = int(player.getAllNBA(playerName2)[1])
    
        
        self.line5b.place( x = axis2, y= 240)
        self.line5b.configure(text= 'All-Defensive Selections: '+player.getAllDefense(playerName2)[1])
        
        player2AllDefense = int(player.getAllDefense(playerName2)[1])
        
        self.line6b.place( x = axis2, y= 270)
        self.line6b.configure(text= 'Championships: '+player.getRings(playerName2)[1])
        
        player2GetRings = int(player.getRings(playerName2)[1])
        
        self.line7b.place( x = axis2, y= 300)
        self.line7b.configure(text= 'Finals MVPs: '+player.getFinalsMVPs(playerName2)[1])
        
        player2FinalsMvps = int(player.getFinalsMVPs(playerName2)[1])
        
        self.line8b.place( x = axis2, y= 330)
        self.line8b.configure(text= 'Regular Season MVPs: '+str(player.getMVPs(playerName2)))
        
        player2Mvp = int(player.getMVPs(playerName2))
        
        self.line9b.place( x = axis2, y= 360)
        self.line9b.configure(text= 'Scoring Titles: '+str(player.getScoringTitles(playerName2)[1]))
        
        player2ScoringTitles = int(player.getScoringTitles(playerName2)[1])
        
        self.goatScore2.place( x = axis2, y= 385)
        self.goatScore2.configure(text= 'GOAT Score: '+str(player.goatScore(playerName2)))
        self.goatScore2.configure(text_color = 'light green')
        self.goatScore2.configure(font =('Roboto',15))
        
        player2GoatScore = int(player.goatScore(playerName2))
        
        player2Metrics = [player2AllStars, player2AllNBA, player2AllDefense, player2GetRings, player2FinalsMvps, player2Mvp, player2ScoringTitles, player2GoatScore ]
        player1Metrics = [player1AllStars, player1AllNBA, player1AllDefense, player1GetRings, player1FinalsMvps, player1Mvp, player1ScoringTitles, player1GoatScore]
        
        #create a green icon if the player 1 stat is greater than the player 2 stat
        icons = self.createIcons()
        
        col = icons[0]
        col2 = icons[1]
        
        i = 0
        pos = 180
        while i < 8:
            if player1Metrics[i]>player2Metrics[i]:
                col[i].place(x = 100, y=pos)
                col2[i].destroy()
            elif player2Metrics[i]>player1Metrics[i]:
                col2[i].place(x=600, y=pos)
                col[i].destroy()
            else:
                col[i].destroy()
                col[i].destroy()
            
            pos+=30
            i+=1

app = App()
app.mainloop()
```
