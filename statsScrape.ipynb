```python
def find(name, fileName):
    f = open(fileName,"r")
    for l in f:
        line = l.split(",")
        if line[0].lower() == name.lower():
            print(line[0].lower())
            return True
        
    return False
      
find('james jone', 'rings.txt')
```

```python
f = open("rings.txt","r")
f2 = open("championships.txt", "r")
new_name = []
rings_names = []
champ_names = []
for l in f:
    line = l.split(",")
    rings_names.append(line[0].lower())
for l in f2:
    line = l.split(",")
    champ_names.append(line[0].lower())      
```

```python
def getRings(name):
    f = open('championships.txt')
    for l in f:
        line = l.split(",")
        if line[0].lower() == name.lower():
            print(line[0].lower()+","+line[1].rstrip("\n"))
            return line[1].rstrip("\n")
        
    return "Not found"
getRings('jackie moore')
```

```python
file = open('rings.txt','w')
for x in new_name:
    rings = getRings(x)
    file.write(x+','+rings)
    file.write("\n")
    
```
