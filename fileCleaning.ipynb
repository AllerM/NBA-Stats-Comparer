```python
f = open("champions.txt", "r")
players= []
for x in f:
    line = x.split("  ")
    players.append(line[0].split("\t"))
```

```python
for x in players:
    print(x.pop(3))

for x in players:
    print(x.pop(2))  
```

```python
with open ("championships.txt", "w") as f:
    for l in players:
        f.write(l[0]+","+l[1])
        f.write("\n")
        
```
